# RFM Metriklerinin Hesaplanması
> Veri setindeki en son alışverişin yapıldığı tarihten 2 gün sonrasını analiz tarihi olarak kabul ediniz.

```python
df["last_order_date"].max() 
```

```python
analysis_date = dt.datetime(2021,6,1)
type(analysis_date)
```

```python
# customer_id, recency, frequnecy ve monetary değerlerinin yer aldığı yeni bir rfm dataframe oluşturma
rfm = pd.DataFrame()
rfm["customer_id"] = df["master_id"]
```

```python
rfm["recency"] = (analysis_date - df["last_order_date"]).dt.days
rfm["recency"][:10]
```

```python
rfm["frequency"] = df["order_num_total"]
```

```python
rfm["monetary"] = df["customer_value_total"]
```

```python
rfm.head()
```

# RF ve RFM Skorlarının Hesaplanması

```python
rfm["recency_score"] = pd.qcut(rfm['recency'], 5, labels=[5, 4, 3, 2, 1])
```

```python
rfm["frequency_score"] = pd.qcut(rfm['frequency'].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
#rank:duplicate önlenmesi
```

```python
rfm["monetary_score"] = pd.qcut(rfm['monetary'], 5, labels=[1, 2, 3, 4, 5])
```

```python
rfm.head()
```

```python
#skor değişkenin oluşturulması
rfm["rf_score"]= (rfm["recency_score"].astype(str) +
                   rfm["frequency_score"].astype(str))
```

```python
rfm["rf_score"][:10]
```

```python
# recency_score ve frequency_score’u tek bir değişken olarak ifade edilmesi ve RF_SCORE olarak kaydedilmesi
rfm["RFM_SCORE"] = (rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str) + rfm['monetary_score'].astype(str))
```

```python
rfm["RFM_SCORE"][:10]
```

# RF Skorlarının Segment Olarak Tanımlanması

```python
seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'
}
```

```python
rfm['segment'] = rfm['rf_score'].replace(seg_map, regex=True)
rfm.head()
```

```python
#Segmentlerin recency, frequnecy ve monetary ortalamalarını inceleyiniz.
rfm[["segment", "recency", "frequency", "monetary"]].groupby("segment").agg(["mean", "count"])
```
